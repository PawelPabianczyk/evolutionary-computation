```python
import pandas as pd
import matplotlib.pyplot as plt

def get_plot(filename, title) -> plt:
    df = pd.read_csv(filename, sep=',')
    x = df.loc[:,"generation"]
    y = df.loc[:,"value"]
    plt.plot(x, y)
    plt.xlabel("generation")
    plt.ylabel("value")
    plt.title(title)
    return plt

get_plot('./20221120134545_best_function_values.csv', "best_function_values").show()
get_plot('./20221120134545_avg_function_values.csv', "avg_function_values").show()
get_plot('./20221120134545_standard_deviations.csv', "standard_deviations").show()
```
